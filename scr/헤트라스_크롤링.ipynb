```python
name=['헤트라스']
category=['별점']

# 헤트라스 디퓨저 후기
ns_address="https://search.shopping.naver.com/catalog/30001171618?query=%EB%94%94%ED%93%A8%EC%A0%80&NaPm=ct%3Dlbauwhsg%7Cci%3D57dba9995117439000e5acdb4d59b19ff932f1c7%7Ctr%3Dslsl%7Csn%3D95694%7Chk%3Dfa3f9374f4e70203a1efb79b8db986213f064a45"
#xpath
shoppingmall_review="/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[1]/ul/li[4]/a"
category1="/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[6]/div[2]/div[2]/ul/li[2]/a" #5점
category2="/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[6]/div[2]/div[2]/ul/li[3]/a" #4점
category3="/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[6]/div[2]/div[2]/ul/li[4]/a" #3점
category4="/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[6]/div[2]/div[2]/ul/li[5]/a" #2점
category5="/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[6]/div[2]/div[2]/ul/li[6]/a" #1점
```

```python
#함수 선언
def add_dataframe(name,category,reviews,stars,dates,options,cnt):  #데이터 프레임에 저장
    #데이터 프레임생성
    df1=pd.DataFrame(columns=['type','category','review','star','date','option'])
    n=1
    if (cnt>0):
        for i in range(0,cnt-1):
            df1.loc[n]=[name,category,reviews[i],stars[i],dates[i],options[i]] #해당 행에 저장
            i+=1
            n+=1
    else:
        df1.loc[n]=[name,category,date,option,'null','null']
        n+=1    
    return df1
```

```python
# 웹페이지 열기

header = {'User-Agent': ''}
d = webdriver.Chrome('/Users/hamsongmin/Codestates/project_2/chromedriver') # webdriver = chrome
d.implicitly_wait(3)
d.get(ns_address)
req = requests.get(ns_address,verify=False)
html = req.text 
soup = BeautifulSoup(html, "html.parser")
sleep(2)

#쇼핑몰 리뷰 보기
d.find_element(By.XPATH, shoppingmall_review).click()
sleep(2)
```

```python
# 5점 리뷰 가져오기
d.find_element(By.XPATH,category1).click() #스크롤 건드리면 안됨
name_=name[0] 
category_=category[0]
reviews=[]
stars=[]
dates=[]
options=[]
cnt=1   #리뷰index
page=1
while True:
    j=1
    print ("페이지", page ,"\n")
    while True: #한페이지에 20개의 리뷰, 마지막 리뷰에서 error발생
        try:
            star=d.find_element(By.XPATH,'/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[6]/ul/li['+str(j)+']/div[1]/span[1]').text
            stars.append(star)
            review=d.find_element(By.XPATH,'/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[6]/ul/li['+str(j)+']/div[2]/div[1]/p').text
            reviews.append(review)
            date=d.find_element(By.XPATH,'/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[6]/ul/li['+str(j)+']/div[1]/span[4]').text
            dates.append(date)
            option=d.find_element(By.XPATH,'/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[6]/ul/li['+str(j)+']/div[1]/span[5]').text
            options.append(option) 
            j+=1
            print(cnt, review, star, date, option, "\n")
            cnt+=1 
        except: break
            
        #sleep(2)
    
    if page<11:#page10까지 적용
        try: #리뷰의 마지막 페이지에서 error발생
            next_page=d.find_element(By.XPATH,'/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[6]/div[3]/a['+str(page)+']').click() 
            page +=1
        except: break #리뷰의 마지막 페이지에서 process 종료 
        
    else :
        try: #page11부터
            if page%10==0: next_page=d.find_element(By.XPATH,'/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[6]/div[3]/a[12]').click()
            else : next_page=d.find_element(By.XPATH,'/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[6]/div[3]/a['+str(page%10+2)+']').click()
            page+=1
        except: break
```

```python
# 5점 리뷰 데이터프레임으로 표현
df1=add_dataframe(name_,category_,reviews,stars,dates,options,cnt)
df1.head()
```

```python
# 크롤링한 5점 데이터의 갯수
len(df1)
```

```python
# 5점 데이터 전처리

df1 = df1.drop_duplicates()
df1 = df1.reset_index(drop=True)
df1_1 = df1['option'].str.split('/',expand=True)
df1 = pd.concat([df1,df1_1],axis=1)
df1.head()
```
